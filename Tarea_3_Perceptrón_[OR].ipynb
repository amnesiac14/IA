```python
import numpy as np
# Entradas para el perceptron
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
# Salidas
Y = np.array([0, 1, 1, 1])
# Pesos para las entradas
weights = np.array([1.3, -2.1])
# Tasa de aprendizaje
lr = 0.001
# Epocas
epochs = 7000
# Sesgo
bias =  4.3
```

Este modelo, el conocido como Perceptrón, introduce formalmente el aprendizaje.
Hasta cierto punto es similar al caso de la neurona de McCullock-Pitts, en el sentido de que tomamos la matriz X conformada por los vectores de entrada (cada renglón de ésta) de cada uno de los casos de la tabla de la operación OR para dos proposiciones.
Calculamos el producto punto con el respectivo vector de pesos, para que dicho resultado entre a la función de activación elegida previamente (en este ejemplo, "el escalón") para devolver, en cada caso de la operación OR, su respectivo resultado.

Aquí entra la diferencia con el modelo neuronal anterior, la cual es la introducción del aprendizaje. En este modelo se introduce el reajuste de los pesos y del sesgo iniciales. Esto se hace para cada uno de los 4 casos del OR, y una vez completados, se termina una epoca.

Este reajuste de cada peso depende de la tasa de aprendizaje -también definida anteriormente-, de la etiqueta o resultado esperado $\vec{y}[i]$ comparándolo con el resultado predicho $\vec{\hat{y}}[i]$, y de cada valor de entrada para cada uno de los casos $i$ de OR

Dicho proceso se repite tantas epocas son necesarias hasta que el modelo consiga los resultados esperados o hasta que se llegue al límite de la capacidad computacional. Sin embargo, es importante aclarar que la comparación $\vec{y}[i]$ con $\vec{\hat{y}}[i]$, que induce una función de pérdida, debería tender a 0 para garantizar que el modelo esté aprendiendo y devolviendo los resultados esperados.

Una vez completado el reajuste, obtenemos los pesos y el sesgo óptimos para poder resolver todos los casos del OR correctamente. Completando el entrenamiento y devolviendo los resultados esperados

```python
class Perceptron:
    def __init__(self, lr, epochs, weights, bias):
        """
            Constructor del perceptron:
            Guarda las variables
            lr -> tasa de aprendizaje
            epochs -> numero de epocas
            weights -> vector de pesos iniciales
            bias -> sesgo inicial
        """
        self.lr = lr
        self.epochs = epochs
        self.weights = weights
        self.bias = bias

    def fit(self, X, Y):
        """
            Realiza el entrenamiento del Perceptron.
        """
        # Recorre el dataset la cantidad indicada en epocas
        for epoch in range(self.epochs):
            for j in range(X.shape[0]):
                # Calcula la salida del perceptrón para la entrada actual
                y_pred = self.activation_function(np.dot(self.weights, X[j]) + self.bias)
                # Calcula el error
                loss = Y[j] - y_pred
                # Actualiza los pesos y el sesgo
                self.weights += self.lr * loss * X[j]
                self.bias += self.lr * loss
            print(f"Epoch {epoch}, Optimized Weights are {self.weights}, and bias is {self.bias}")
        # Imprime los valores finales de los parámetros aprendidos
        print(f"Optimized Weights are {self.weights} and bias is {self.bias}")

    def activation_function(self, activation):
        """
            Función de activacion escalon
        """
        return 1 if activation >= 0 else 0

    def prediction(self, X):
        """
            Calcula la salida del Perceptron para cada fila de entradas X.
        """
        # Calcula producto punto + bias para todas las entradas
        sum_ = np.dot(X, self.weights) + self.bias
        # Mensaje input y su predicción
        for i, s in enumerate(sum_):
            print(f"Input: {X[i]}, Predictions: {self.activation_function(sum_[i])}")
        # Devuelve un array con todas las predicciones
        return np.array([self.activation_function(s) for s in sum_])
```

```python
# Crear una instancia del perceptrón
p = Perceptron(lr=lr, epochs=epochs, weights=weights, bias=bias)
# Entrenar el modelo
p.fit(X, Y)
# Usar el modelo entrenado para realizar predicciones
predictions = p.prediction(X)
```
